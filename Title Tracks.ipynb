```python
import numpy as np
import pandas as pd
```

```python
# https://www.kaggle.com/rodolfofigueroa/spotify-12m-songs
data = pd.read_csv('tracks_features.csv')
```

```python
data = data[['name', 'album', 'album_id', 'artists', 'track_number', 'disc_number']]
```

```python
data
```

```python
data.rename(columns={'name': 'song'}, inplace=True)
```

```python
data['album_id'].unique().size #118382 total albums
```

```python
data['album'].value_counts().head(60)
```

```python
data['album_id'].value_counts() # Greatest hits compilations, singles, EPs?
```

```python
album_lengths = data.groupby('album_id').agg({'track_number':['max']})
```

```python
album_lengths.value_counts() # Singles are the most common album, 20994 of them
```

```python
# Focus on albums between 8 and 16 tracks
LP_ids = album_lengths[album_lengths[('track_number', 'max')].isin(range(8,17))].index # still a lot of live albums and stuff
```

```python
LP_ids.size # 63227 for 8 to 16 length albums, 64717 if ignoring disc 2+
```

```python
# Just deal with it I guess
LPs = data.loc[data['album_id'].isin(LP_ids)]
```

```python
LPs.groupby('album_id').agg({'track_number':['max']}).value_counts()
```

```python
#title_tracks = LPs.loc[LPs['name'] == LPs['album']]
title_tracks = LPs[LPs.apply(lambda x: x.album in x.song or x.song in x.album, axis=1)] # little slow
```

```python
title_tracks #17303 for exact match #20421 for song IN album #23636 for album IN song #26754 for either
```

```python
title_tracks['track_number'].value_counts()
```

```python
title_tracks.to_csv('title_tracks.csv')
```
